```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Iron_Ore.csv')
```

```python
data.head()
```

```python
data = data.drop(['LATITUDE','LONGITUDE', 'COMMODITY'], axis = 1)
```

```python
data.isna().sum()
```

```python
data['METALLOGENESIS'].value_counts()
```

```python
data["METALLOGENESIS"]=data["METALLOGENESIS"].str.replace('RAMGIRI-P NAKACHERLA-HUNGUD BELT','RAMAGIRI-PENAKACHERLA-HUNGUD BELT',regex=False)
data["METALLOGENESIS"]=data["METALLOGENESIS"].str.replace('RAMGIRI-P AKACHERLA-HUNGUD BELT\n','RAMAGIRI-PENAKACHERLA-HUNGUD BELT',regex=False)
data["METALLOGENESIS"]=data["METALLOGENESIS"].str.replace('NEB RANGE(SANDUR BELT)','SANDUR SCHIST BELT-NEB RANGE',regex=False)
data["METALLOGENESIS"]=data["METALLOGENESIS"].str.replace('NEB RANGE (SANDUR SCHIST BELT)','SANDUR SCHIST BELT-NEB RANGE',regex=False)
data["METALLOGENESIS"]=data["METALLOGENESIS"].str.replace('BONAI-NOAMUNDI-JAMDA PROVINCE','BONAI-NOAMUNDI-JAMDA BELT',regex=False)
data["METALLOGENESIS"]=data["METALLOGENESIS"].str.replace('WURRYA HILL RANGE(?)','WURRYA HILL RANGE',regex=False)
data["METALLOGENESIS"]=data["METALLOGENESIS"].str.replace('PUR BANERA BELT(EASTERN ZONE)','PUR-BANERA BELT(EASTERN ZONE)',regex=False)
data["METALLOGENESIS"]=data["METALLOGENESIS"].str.replace('RAMANDUR RANGE(SHIMOGA SCHIST) BELT','SHIMOGA SCHIST BELT(RAMANDUR RANGE)',regex=False)
data["METALLOGENESIS"]=data["METALLOGENESIS"].str.replace('KAPPALGUDDA (SHIMOGA SCHIST) BELT','SHIMOGA SCHIST BELT(KAPPALGUDDA RANGE)',regex=False)
data["METALLOGENESIS"]=data["METALLOGENESIS"].str.replace('PUR-BANERA BELT (WESTERN ZONE)','PUR-BANERA BELT(WESTERN ZONE)',regex=False)
data["METALLOGENESIS"]=data["METALLOGENESIS"].str.replace('NEB(SANDUR SCHIST BELT)','SANDUR SCHIST BELT-NEB RANGE',regex=False)
data["METALLOGENESIS"]=data["METALLOGENESIS"].str.replace('NEB (SANDUR SCHIST BELT)','SANDUR SCHIST BELT-NEB RANGE',regex=False)
data["METALLOGENESIS"]=data["METALLOGENESIS"].str.replace('BELLARY (SANDUR SCHIST BELT)','SANDUR SCHIST BELT-BELLARY',regex=False)
data["METALLOGENESIS"]=data["METALLOGENESIS"].str.replace('SANDUR BELT-BELLARY SECTOR','SANDUR SCHIST BELT-BELLARY',regex=False)
```

```python
print(data['METALLOGENESIS'].value_counts().to_string())
```

```python
print(data['LOCALITY'].value_counts().to_string())
```

```python
print(data['STATE'].value_counts().to_string())
```

```python
data["STATE"]=data["STATE"].str.replace('ORISSA (?)','ORISSA',regex=False)
data["STATE"]=data["STATE"].str.replace('KARNATAKA, A.P','KARNATAKA',regex=False)
```

```python
print(data['TOPOSHEET'].value_counts().to_string())
```

```python
data['TOPOSHEET'] = data['TOPOSHEET'].str.replace('57B','57 B', regex = False)
```

```python
from pandas_profiling import ProfileReport
profile = ProfileReport(data, title="Iron Ore Report", explorative=True)
```

```python
profile.to_file("iron-report.html")
```

```python
map_india = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(data['LATDD'], data['LONDD'], data['NEW']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_india)  
```
