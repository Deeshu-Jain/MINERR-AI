```python
data = pd.read_csv("Lead-Zinc_Ore_deposits_in_India_and_its_salient_features.csv")
```

```python
data.head()
```

```python
from geopy.geocoders import Nominatim
import folium
import re
import json
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe
from sklearn.cluster import KMeans
```

```python
latitude = 24.95000
longitude = 74.133333
```

```python
data["NEW"] = data["commodity"]+", "+data["locality"]
```

```python
data["NEW"]
```

```python
map_india = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(data['Y'], data['X'], data['NEW']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_india)  
```
