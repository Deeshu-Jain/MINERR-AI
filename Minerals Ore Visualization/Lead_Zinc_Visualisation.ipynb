```python
from geopy.geocoders import Nominatim
import folium
import re
import json
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe
from sklearn.cluster import KMeans
```

```python
latitude = 24.95000
longitude = 74.133333
```

```python
data["NEW"] = data["commodity"]+", "+data["locality"]
```

```python
data["NEW"]
```

```python
map_india = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(data['Y'], data['X'], data['NEW']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_india)  
```

```python
map_html = map_india._repr_html_()
```

```python
print(map_html)
```

```python
data.head()
```

```python
data["statename"].value_counts()
```

```python
features = data[['Y', 'X']]
print(features)
```

```python
from sklearn.cluster import KMeans
# create kmeans model/object
kmeans = KMeans(
    init="random",
    n_clusters=4,
    n_init=10,
    max_iter=300,
    random_state=42
)
```

```python
kmeans.fit(features)
# save results
labels = kmeans.labels_
```

```python
data['cluster'] = labels
```

```python
data['LOC']=data["Y"].astype(str) +", "+data["X"].astype(str)
```

```python
# display the number of mamber each clustering
_clusters = data.groupby('cluster')["statename"].count()
print(_clusters)
```

```python
colors = ['red', 'blue', 'green', 'purple']
lat = 28.65
lng = 77.23

map = folium.Map(location=[lng, lat], zoom_start=12)
for lat, lng, cluster,label in zip(data['Y'], data['X'], data['cluster'],data['LOC']):
# for _, row in data.iterrows():
    label = folium.Popup(label,parse_html=True)
    folium.CircleMarker(
        location=[lat, lng],
        popup=label,
        radius=15, 
        weight=2, 
        fill=True, 
        fill_color=colors[cluster],
        color=colors[cluster]
    ).add_to(map)
map
```
